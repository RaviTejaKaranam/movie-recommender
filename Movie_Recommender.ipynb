```python
movie_id_name.info()
```

```python
movie_rating = pd.read_csv("Downloads/the-movies-dataset/ratings.csv")
movie_rating.drop("timestamp",axis=1,inplace=True)
movie_rating.columns = ["UserId",'id','rating']
df = pd.merge(movie_id_name,movie_rating,on='id')
df.head()
```

```python
df.groupby("original_title")['rating'].mean().sort_values(ascending=False)
df.groupby("original_title")['rating'].count().sort_values(ascending=False)
```

```python
ratings = pd.DataFrame(df.groupby("original_title")['rating'].mean())
ratings.head()
```

```python
ratings['num of ratings']=pd.DataFrame(df.groupby('original_title')['rating'].count())
```

```python
movie_table = df.pivot_table(index="UserId",columns='original_title',values='rating')
movie_table.info()
```

```python
ratings.sort_values('num of ratings',ascending=False).head()
```

```python
EVA_user_ratings = movie_table['EVA']
```

```python
movie_table.corrwith(EVA_user_ratings)
```

```python
similar_to_EVA = movie_table.corrwith(EVA_user_ratings)
```

```python
corr_EVA = pd.DataFrame(similar_to_EVA,columns = ['Correlation'])
```

```python
corr_EVA.dropna(inplace=True)
```

```python
corr_EVA = corr_EVA.join(ratings['num of ratings'])
```

```python
corr_EVA.head()
```

```python
corr_EVA[corr_EVA['num of ratings']>50].sort_values('Correlation',ascending=False)
```
